## Date Formatting

### Subtask:
Convert the 'publish_time' and 'trending_date' columns into datetime objects to enable time-based calculations.


**Reasoning**:
I will convert the 'publish_time' and 'trending_date' columns to datetime objects using appropriate formats to ensure accurate time-based analysis as requested in the subtask instructions.



```python
import pandas as pd

# Convert 'publish_time' to datetime objects
# Standard ISO 8601 format usually handled automatically, but we ensure UTC awareness/handling
df['publish_time'] = pd.to_datetime(df['publish_time'])

# Convert 'trending_date' to datetime objects using the specific format YY.DD.MM
df['trending_date'] = pd.to_datetime(df['trending_date'], format='%y.%d.%m')

# Verify the conversion
print("Data types after conversion:")
print(df[['publish_time', 'trending_date']].dtypes)

# Display the first few rows to verify the values
display(df[['publish_time', 'trending_date']].head())
```

## Feature Engineering

### Subtask:
Calculate the difference between 'trending_date' and 'publish_time' in hours and store it in a new column called 'time_to_trend'.


**Reasoning**:
I will calculate the time difference in hours between 'trending_date' and 'publish_time' by handling the timezone mismatch and storing the result in a new column 'time_to_trend'.



```python
import pandas as pd

# Ensure publish_time is naive to match trending_date for subtraction
df['publish_time_naive'] = df['publish_time'].dt.tz_localize(None)

# Calculate the difference and convert to hours
# Using total_seconds() / 3600 for precision
df['time_to_trend'] = (df['trending_date'] - df['publish_time_naive']).dt.total_seconds() / 3600

# Drop the temporary naive column
df.drop(columns=['publish_time_naive'], inplace=True)

# Display the results
print("Time to trend calculation complete.")
display(df[['publish_time', 'trending_date', 'time_to_trend']].head())

# Check basic statistics of the new column
print("\nSummary statistics for 'time_to_trend':")
print(df['time_to_trend'].describe())
```

## Calculate Engagement Rate

### Subtask:
Engineer a new feature called 'engagement_rate' calculated as the sum of 'likes' and 'comment_count' divided by 'views'.


**Reasoning**:
I will engineer the 'engagement_rate' feature by summing 'likes' and 'comment_count' and dividing by 'views', while safely handling potential division by zero using a conditional check.



```python
import numpy as np

# Calculate engagement rate: (likes + comment_count) / views
# Using np.where to handle cases where views might be 0 to avoid division by zero
df['engagement_rate'] = np.where(
    df['views'] > 0,
    (df['likes'] + df['comment_count']) / df['views'],
    0
)

# Display the first few rows of the updated DataFrame for verification
print("Engagement rate feature engineered.")
display(df[['views', 'likes', 'comment_count', 'engagement_rate']].head())

# Check the distribution of the new column
print("\nBasic statistics for engagement_rate:")
print(df['engagement_rate'].describe())
```

**Reasoning**:
I will visualize the distribution of the newly created 'engagement_rate' using a histogram and a Kernel Density Estimate (KDE) plot to understand its spread and identify any common engagement patterns across the dataset.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the visual style
sns.set_style('whitegrid')

# Initialize the figure
plt.figure(figsize=(12, 6))

# Plot the distribution of engagement_rate
sns.histplot(df['engagement_rate'], bins=50, kde=True, color='teal')

# Add titles and labels
plt.title('Distribution of Engagement Rate', fontsize=15)
plt.xlabel('Engagement Rate (Likes + Comments / Views)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

# Adjust x-axis limits if there are extreme outliers to make the bulk of the data visible
# Based on statistics, most data is under 0.1
plt.xlim(0, 0.2)

plt.show()
```

# 7. K-Means Clustering

# Task
Perform K-Means clustering on the "/content/INvideos.csv" dataset to segment videos based on engagement by following these steps:
1. **Prepare and Scale Features**: Select the 'views', 'likes', and 'comment_count' columns and normalize them using `StandardScaler` to ensure the clustering algorithm treats them equally.
2. **Determine Optimal Clusters**: Use the Elbow Method by calculating and plotting the Within-Cluster Sum of Squares (WCSS) for 1 to 10 clusters to identify the ideal number of groupings.
3. **Apply K-Means**: Train the KMeans model with the optimal cluster count, fit it to the scaled data, and append the resulting cluster labels to your original DataFrame.
4. **Visualize Results**: Generate a scatter plot (e.g., Views vs. Likes) colored by the cluster labels to visualize how the videos are segmented.
5. **Summarize Findings**: Describe the characteristics of each identified cluster (e.g., distinguishing between high-engagement viral hits and low-engagement videos) and analyze the distribution of videos across these groups.

## Prepare and Scale Features

### Subtask:
Select the 'views', 'likes', and 'comment_count' columns from the existing DataFrame and normalize them using `StandardScaler` for clustering.


**Reasoning**:
I will select the required numerical columns, import StandardScaler, and perform normalization to prepare the data for clustering.



```python
from sklearn.preprocessing import StandardScaler

# 1. Select the relevant features
clustering_subset = df[['views', 'likes', 'comment_count']]

# 2. Instantiate and apply StandardScaler
scaler = StandardScaler()
scaled_features = scaler.fit_transform(clustering_subset)

# 3. Print verification info
print(f"Shape of scaled_features: {scaled_features.shape}")
print("\nFirst 5 rows of original subset for verification:")
display(clustering_subset.head())

# 4. Show a sample of the scaled data
print("\nFirst 2 rows of scaled data:")
print(scaled_features[:2])
```

## Elbow Method for Optimal Clusters

### Subtask:
Iterate through a range of cluster counts (1 to 10) and calculate the Within-Cluster Sum of Squares (WCSS) to identify the optimal number of clusters.


**Reasoning**:
I will perform the Elbow Method by iterating through cluster counts from 1 to 10, calculating the Within-Cluster Sum of Squares (WCSS) using KMeans, and plotting the results to identify the optimal number of clusters.



```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# 1. Initialize an empty list to store the inertia values (WCSS)
wcss = []

# 2. Iterate through a range of cluster counts (1 to 10)
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(scaled_features)
    wcss.append(kmeans.inertia_)

# 3. Plot the Elbow Method results
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), wcss, marker='o', linestyle='--', color='blue')

# 4. Add labels and title
plt.title('Elbow Method for Optimal Clusters', fontsize=15)
plt.xlabel('Number of Clusters', fontsize=12)
plt.ylabel('WCSS (Inertia)', fontsize=12)
plt.xticks(range(1, 11))
plt.grid(True)

# 5. Display the plot
plt.show()

print("WCSS values per cluster count:")
for i, val in enumerate(wcss, 1):
    print(f"Clusters {i}: {val:.2f}")
```

## Apply K-Means Clustering

### Subtask:
Train the KMeans model with the optimal cluster count, fit it to the scaled data, and append the resulting cluster labels to the original DataFrame.


**Reasoning**:
I will train the KMeans model with the optimal cluster count of 4, fit it to the scaled features, and append the resulting labels to the DataFrame 'df' under a new 'cluster' column as specified in the instructions.



```python
from sklearn.cluster import KMeans

# 1. Instantiate the KMeans model with n_clusters=4
kmeans = KMeans(n_clusters=4, init='k-means++', random_state=42)

# 2. Fit the model to the scaled_features and predict cluster labels
df['cluster'] = kmeans.fit_predict(scaled_features)

# 3. Display the first few rows focusing on engagement metrics and the new cluster column
print("K-Means clustering complete. Cluster labels assigned.")
display(df[['views', 'likes', 'comment_count', 'cluster']].head())

# 4. Check the distribution of videos across clusters
print("\nNumber of videos per cluster:")
print(df['cluster'].value_counts())
```

## Visualize Clusters

### Subtask:
Create a scatter plot to visualize the K-Means clustering results based on video engagement.


**Reasoning**:
I will create a scatter plot using seaborn to visualize the results of the K-Means clustering, plotting views against likes and coloring the points by their assigned cluster label to show the engagement segments.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the visual style
sns.set_style('whitegrid')

# Initialize the figure
plt.figure(figsize=(12, 8))

# Create the scatter plot
# x='views', y='likes', colored by 'cluster'
sns.scatterplot(data=df, x='views', y='likes', hue='cluster', palette='viridis', alpha=0.7, s=60)

# Add title and clear labels
plt.title('Video Engagement Clusters: Views vs. Likes', fontsize=16)
plt.xlabel('Total Views', fontsize=12)
plt.ylabel('Total Likes', fontsize=12)

# Add a legend to explain the clusters
plt.legend(title='Cluster ID', loc='upper left')

# Display the plot
plt.show()
```
